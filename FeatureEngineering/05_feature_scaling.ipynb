```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Age": [20, 25, 30, 35, 40],
    "Fare": [10, 50, 20, 100, 80]
})
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.plot(df["Age"], label="Age")
plt.plot(df["Fare"], label="Fare")
plt.title("Before Scaling")
plt.legend()
plt.show()
```

### 1. Standard Scalar

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

st_scaled = scaler.fit_transform(df)
st_scaled = pd.DataFrame(st_scaled, columns=df.columns)
st_scaled
```

```python
plt.figure(figsize=(6,4))
plt.plot(st_scaled["Age"], label="Age (scaled)")
plt.plot(st_scaled["Fare"], label="Fare (scaled)")
plt.title("After Standard Scaling")
plt.legend()
plt.show()
```

### 2. Roust Scalar

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
rb_scaled = scaler.fit_transform(df)
rb_scaled = pd.DataFrame(rb_scaled, columns=df.columns)
rb_scaled
```

```python
plt.figure(figsize=(6,4))
plt.plot(rb_scaled["Age"], label="Age (scaled)")
plt.plot(rb_scaled["Fare"], label="Fare (scaled)")
plt.title("After Robust Scaling")
plt.legend()
plt.show()
```

### 3. MinMax Scalar

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
mm_scaled = scaler.fit_transform(df)
mm_scaled = pd.DataFrame(mm_scaled, columns=df.columns)
mm_scaled
```

```python
plt.figure(figsize=(6,4))
plt.plot(mm_scaled["Age"], label="Age (scaled)")
plt.plot(mm_scaled["Fare"], label="Fare (scaled)")
plt.title("After Robust Scaling")
plt.legend()
plt.show()
```
