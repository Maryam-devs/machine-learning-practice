```python
import pandas as pd
```

```python
# Dummy Dataset
df = pd.DataFrame({
    "Education": ["High School", "Bachelor's", "High School", "Master's", "PhD", "High School", "Bachelor's", "Master's", "PhD", "Bachelor's", "High School"]
})

df["Education"].value_counts()
```

### 1. Label Encoding

```python
le_df = df.copy()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

le_df["Education"] = le.fit_transform(le_df["Education"])
```

```python
le_df["Education"]
```

### 2. One Hot Encoding

### Method 1 (Using Pandas)

```python
ohe_df1 = df.copy()
```

```python
df1_encoded = pd.get_dummies(ohe_df1, columns=["Education"])
df1_encoded
```

### Method 2 (Using sklearn)

```python
ohe_df2 = df.copy()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)

encoded = encoder.fit_transform(ohe_df2[["Education"]])

encoded
```

```python
encoded_df = pd.DataFrame(
    encoded,
    columns=encoder.get_feature_names_out(["Education"])
)

encoded_df
```

### 3. Ordinal Encoding

```python
oe_df = df.copy()
```

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder(
    categories=[["High School", "Bachelor's", "Master's", "PhD"]]
)

oe_df["Edu_encoded"] = encoder.fit_transform(oe_df[["Education"]])
oe_df
```

### 4. Frequency Encoding

```python
fe_df = df.copy()
```

```python
freq = fe_df["Education"].value_counts(normalize= True)
freq
```

```python
fe_df["Edu_freq"] = fe_df["Education"].map(freq)
fe_df
```

### 5. Count Encoding

```python
ce_df = df.copy()
```

```python
count = ce_df["Education"].value_counts()
count
```

```python
ce_df["Edu_count"] = ce_df["Education"].map(count)
ce_df
```
